## Import libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Use this plot style

```python
plt.style.use('seaborn')
```

## Import csv file

```python
data = pd.read_csv('F://Files/Python work/case study/2019-05-31-data.csv')
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

## Here is a look at a scatter plot

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer', 
           edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
