## Import libraries

```python
import csv
import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

## Use this plot style

```python
plt.style.use('fivethirtyeight')
```

## Import csv file

```python
data = pd.read_csv('F://Files/Python work/case study/data1.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
```

## Another way

```python
with open('F://Files/Python work/case study/data1.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
print(language_counter.most_common(15))
```

## Counter

```python
language_counter = Counter()
```

## For loop for updating

```python
for response in lang_responses:
    language_counter.update(response.split(';'))
```

```python
languages = []
popularity = []
```

```python
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)
```

## Early Plot

```python
plt.bar(languages, popularity)

plt.title('Most Popular Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Number of People who Use')
```

## How to use a horizontal bar (barh)

```python
plt.barh(languages, popularity)

plt.title('Most Popular Languages')
# plt.xlabel('Programming Languages')
plt.xlabel('Number of People who Use')
```

## Reverse the order where the most popular is on top

```python
languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title('Most Popular Languages')
# plt.xlabel('Programming Languages')
plt.xlabel('Number of People who Use')
```
