## Import libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Import data as csv file

```python
data = pd.read_csv('F://Files/Python work/case study/data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

## Early plot

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
```

## Plotting 'Python'

```python
plt.plot(ages, py_salaries, label='Python')
```

```python
overall_median = 57287
```

## Combining all plots

```python
overall_median = 57287

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25) # lighter shade of blue (alpha)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25) # lighter shade of blue (alpha)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25) # lighter shade of blue (alpha)

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries < overall_median), interpolate=True, color='red', alpha=0.25) # lighter shade of blue (alpha)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate=True, alpha=0.25, label='Above avg') # lighter shade of blue (alpha)

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), interpolate=True, color='red', alpha=0.25, label='Below avg') # lighter shade of blue (alpha)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```
